```python

python_df = survey_df[survey_df["LanguageHaveWorkedWith"].str.contains("Python", na=False)]
python_df = python_df[["Country", "ConvertedCompYearly"]].dropna()

comp_by_country = (
    python_df
    .groupby("Country")["ConvertedCompYearly"]
    .agg(mean_compensation="mean",
        median_compensation="median")
    .round(2)  # округлення до 2 знаків після коми
    .sort_values("mean_compensation", ascending=False))

comp_by_country.head(20)

```

```python
age_python = survey_df.copy()
age_python["UsesPython"] = age_python["LanguageHaveWorkedWith"].str.contains("Python", na=False)

python_by_age = (
    age_python
    .groupby("Age")
    .agg(
        total=("UsesPython", "count"),
        python_users=("UsesPython", "sum")
    )
)

python_by_age["python_percentage"] = round(
    python_by_age["python_users"] / python_by_age["total"] * 100, 2
)

python_by_age[["python_percentage"]].dropna()
```
